```python
#County <-> Votes
import numpy as np
from sklearn import linear_model
import pandas as pd
```

```python
df = pd.read_csv("votes-train.csv")
df.head(20)
```

```python
ytr = np.array(df['Democrat'])
xtr = np.column_stack((np.array(df['Black']),np.array(df['Edu_bachelors'])))
#xtr = np.column_stack((np.array(df['population']),np.array(df['Edu_bachelors'])))
```

```python
logreg = linear_model.LogisticRegression(C=1e5)
logreg.fit(xtr, ytr)
```

```python
yhat = logreg.predict(xtr)
accuracy = list(ytr==yhat).count(True)/len(ytr)
print(accuracy)
```

```python
df2 = pd.read_csv("votes-test.csv")
df2.head(6)
yts = np.array(df2['Democrat'])
xts = np.column_stack((np.array(df2['Black']),np.array(df2['Edu_bachelors'])))
#xts = np.column_stack((np.array(df2['population']),np.array(df2['Edu_bachelors'])))
```

```python
yhat = logreg.predict(xts)
accuracy = list(yts==yhat).count(True)/len(yts)
print(accuracy)
```
